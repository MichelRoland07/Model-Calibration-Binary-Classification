```python
# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_val_scaled = scaler.transform(X_val)

# List of models to train with specific variable names
svm_model = SVC(kernel="linear", C=5.803122849969755, gamma=0.03906939937054617)
rf_model = RandomForestClassifier(random_state=123)
lr_model = LogisticRegression(random_state=123)
nb_model = GaussianNB()
xg_model= XGBClassifier()


models = {
    "SVM Model": svm_model,
    "Random Forest Model": rf_model,
    "Logistic Regression Model": lr_model,
    "XGBoost Model": xg_model,
    "Naive Bayes Model": nb_model,
    
}

# Function to train and evaluate models
def train_(model, X_train, y_train) :
    # Fit the model
    model.fit(X_train, y_train)

    # Loop over each model and train/evaluate
for model_name, model in models.items():
    print(f"Training {model_name}...")
    train_(model, X_train, y_train)

```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import make_scorer, f1_score

# Define hyperparameter search space for each model
param_grids = {
    "SVM Model": {
        "kernel": ["linear", "rbf"],
        "C": [0.1, 1, 10, 100],
        "gamma": [0.001, 0.01, 0.1, 1]
    },
    "Random Forest Model": {
        "n_estimators": [50, 100, 200],
        "max_depth": [None, 10, 20, 30],
        "min_samples_split": [2, 5, 10],
        "min_samples_leaf": [1, 2, 4]
    },
    "Logistic Regression Model": {
        "penalty": ["l2"],
        "C": [0.1, 1, 10, 100],
        "solver": ["lbfgs", "saga"]
    },
    "XGBoost Model": {
        "learning_rate": [0.01, 0.1, 0.2],
        "n_estimators": [50, 100, 200],
        "max_depth": [3, 5, 7],
        "gamma": [0, 0.1, 0.2],
        "subsample": [0.8, 1.0]
    },
    "Naive Bayes Model": {}  # GaussianNB doesn't have tunable hyperparameters
}

# Metric for optimization
scoring_metric = make_scorer(f1_score, average="binary")

# Function to perform hyperparameter tuning
def tune_hyperparameters(model, param_grid, X_train, y_train, model_name):
    if param_grid:  # Only tune if there are hyperparameters to tune
        print(f"Tuning {model_name}...")
        grid_search = GridSearchCV(
            estimator=model,
            param_grid=param_grid,
            scoring=scoring_metric,
            cv=5,
            verbose=1,
            n_jobs=-1
        )
        grid_search.fit(X_train, y_train)
        print(f"Best parameters for {model_name}: {grid_search.best_params_}")
        print(f"Best F1-score for {model_name}: {grid_search.best_score_:.4f}")
        return grid_search.best_estimator_
    else:
        print(f"No hyperparameters to tune for {model_name}.")
        model.fit(X_train, y_train)
        return model

# Loop over each model, perform tuning, and train with best hyperparameters
tuned_models = {}
for model_name, model in models.items():
    best_model = tune_hyperparameters(model, param_grids.get(model_name, {}), X_train_scaled, y_train, model_name)
    tuned_models[model_name] = best_model
```

<font color='blue' size="+2">Metric Justification: F1-Score</font>

Given the imbalanced nature of the dataset, accuracy alone may be misleading. The F1-score, as the harmonic mean of precision and recall, is better suited for evaluating model performance when dealing with imbalanced data. It ensures that both false positives and false negatives are accounted for, providing a more comprehensive assessment.

 <font color='blue' size="+3">Calibration Techniques</font>

<font color='blue' size="+2">Implementation and metrics evaluation</font>

```python


# Beta Calibration Class
def log_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

class BetaCalibration:
    def __init__(self):
        self.alpha = None
        self.beta = None

    def fit(self, probs, y):
        # Log-loss for beta calibration
        def beta_loss(params):
            alpha, beta_param = params
            cal_probs = beta.cdf(probs, alpha, beta_param)
            return log_loss(y, cal_probs)

        # Optimize alpha and beta
        result = minimize(beta_loss, x0=[1.0, 1.0], bounds=[(0.01, 10), (0.01, 10)])
        self.alpha, self.beta = result.x

    def transform(self, probs):
        return beta.cdf(probs, self.alpha, self.beta)

# Define classifiers
clf_dict = {
    "SVM Model": svm_model_t,
    "Random Forest Model": rf_model_t,
    "Logistic Regression Model": lr_model_t,
    "XGBoost Model": xg_model_t,
    "Naive Bayes Model": nb_model_t,
}

calibrated_probs = {}

# Function to calculate Expected Calibration Error (ECE)
def calculate_ece(probs, y_true, bins=10):
    ece = 0
    for bin_start, bin_end in zip(np.linspace(0, 1, bins)[:-1], np.linspace(0, 1, bins)[1:]):
        bin_mask = (probs >= bin_start) & (probs < bin_end)
        bin_true = y_true[bin_mask]
        bin_pred = probs[bin_mask]
        if len(bin_pred) > 0:
            ece += np.abs(np.mean(bin_pred) - np.mean(bin_true)) * len(bin_true) / len(y_true)
    return ece

# Loop over classifiers
for model_name, clf in clf_dict.items():
    print(f"\nFitting {model_name}...")

    # Train the classifier
    clf.fit(X_train, y_train)

    # Get the predicted probabilities for the validation set
    if hasattr(clf, "predict_proba"):
        prob_pos_val = clf.predict_proba(X_val)[:, 1]
        prob_pos_test = clf.predict_proba(X_test)[:, 1]
    else:
        # For models without predict_proba, normalize decision_function outputs to [0, 1]
        prob_pos_val = clf.decision_function(X_val)
        prob_pos_val = (prob_pos_val - prob_pos_val.min()) / (prob_pos_val.max() - prob_pos_val.min())
        prob_pos_test = clf.decision_function(X_test)
        prob_pos_test = (prob_pos_test - prob_pos_test.min()) / (prob_pos_test.max() - prob_pos_test.min())

    # Apply Platt Scaling
    platt_calibrator = CalibratedClassifierCV(clf, method='sigmoid', cv='prefit')
    platt_calibrator.fit(X_val, y_val)
    y_pred_probs_platt = platt_calibrator.predict_proba(X_test)[:, 1]
    brier_platt = brier_score_loss(y_test, y_pred_probs_platt)
    ece_platt = calculate_ece(y_pred_probs_platt, y_test)
    print(f"Brier score after Platt Scaling for {model_name}: {brier_platt:.4f}")
    print(f"Expected Calibration Error (ECE) after Platt Scaling for {model_name}: {ece_platt:.4f}")

    # Plot Reliability Diagram for Platt Scaling
    prob_true_platt, prob_pred_platt = calibration_curve(y_test, y_pred_probs_platt, n_bins=10, strategy='uniform')
    plt.figure()
    plt.plot(prob_pred_platt, prob_true_platt, marker='o', label=f"{model_name} (Platt Scaling)")
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label="Perfect Calibration")
    plt.xlabel("Predicted Probability")
    plt.ylabel("True Probability")
    plt.title(f"Reliability Diagram - {model_name} (Platt Scaling)")
    plt.legend()
    plt.grid()
    plt.show()

    # Apply Isotonic Regression
    isotonic_calibrator = CalibratedClassifierCV(clf, method='isotonic', cv='prefit')
    isotonic_calibrator.fit(X_val_scaled, y_val)
    y_pred_probs_isotonic = isotonic_calibrator.predict_proba(X_test_scaled)[:, 1]
    brier_isotonic = brier_score_loss(y_test, y_pred_probs_isotonic)
    ece_isotonic = calculate_ece(y_pred_probs_isotonic, y_test)
    print(f"Brier score after Isotonic Regression for {model_name}: {brier_isotonic:.4f}")
    print(f"Expected Calibration Error (ECE) after Isotonic Regression for {model_name}: {ece_isotonic:.4f}")

    # Plot Reliability Diagram for Isotonic Regression
    prob_true_isotonic, prob_pred_isotonic = calibration_curve(y_test, y_pred_probs_isotonic, n_bins=10, strategy='uniform')
    plt.figure()
    plt.plot(prob_pred_isotonic, prob_true_isotonic, marker='o', label=f"{model_name} (Isotonic Regression)")
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label="Perfect Calibration")
    plt.xlabel("Predicted Probability")
    plt.ylabel("True Probability")
    plt.title(f"Reliability Diagram - {model_name} (Isotonic Regression)")
    plt.legend()
    plt.grid()
    plt.show()

    # Apply Beta Calibration
    bc = BetaCalibration()
    bc.fit(prob_pos_val, y_val)
    calibrated_test_probs = bc.transform(prob_pos_test)
    brier_beta = brier_score_loss(y_test, calibrated_test_probs)
    ece_beta = calculate_ece(calibrated_test_probs, y_test)
    print(f"Brier score after Beta Calibration for {model_name}: {brier_beta:.4f}")
    print(f"Expected Calibration Error (ECE) after Beta Calibration for {model_name}: {ece_beta:.4f}")

    # Store calibrated probabilities
    calibrated_probs[model_name] = calibrated_test_probs

    # Evaluate Beta Calibration
    print(f"\nEvaluation for {model_name} with Beta Calibration:")
    print(f"Brier Score: {brier_score_loss(y_test, calibrated_test_probs):.4f}")
    print(f"Log Loss: {log_loss(y_test, calibrated_test_probs):.4f}")
    print(classification_report(y_test, (calibrated_test_probs > 0.5).astype(int)))

    # Plot Reliability Diagram for Beta Calibration
    prob_true, prob_pred = calibration_curve(y_test, calibrated_test_probs, n_bins=10, strategy='uniform')
    plt.figure()
    plt.plot(prob_pred, prob_true, marker='o', label=f"{model_name} (Beta Calibrated)")
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label="Perfect Calibration")
    plt.xlabel("Predicted Probability")
    plt.ylabel("True Probability")
    plt.title(f"Reliability Diagram - {model_name} (Beta Calibration)")
    plt.legend()
    plt.grid()
    plt.show()
```
